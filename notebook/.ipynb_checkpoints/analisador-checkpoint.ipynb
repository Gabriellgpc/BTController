```python
motor_e = pd.read_csv('../etc/motor_esquerdo.csv')
motor_d = pd.read_csv('../etc/motor_esquerdo.csv')
```

```python
motor_e.head()
```

```python
t = motor_e['TIME']
we= motor_e['OMEGA_RAW']
wd= motor_d['OMEGA_RAW']

plt.plot(t, we, '-b', label=r'$\omega_e$')
plt.plot(t, wd, '-g', label=r'$\omega_d$')

plt.ylabel(r'$rad.s^{-1}$')
plt.xlabel(r'$t$[s]')
plt.grid(True)
plt.legend()
plt.show()
```
